```python

# Drop identifiers not useful for recommendation
#df = df.drop(columns=["InvoiceNo"], errors="ignore")
import pickle
# Create an Item-Customer matrix (rows = products, columns = customers)
item_customer_matrix = df.pivot_table(
    index='Description',
    columns='CustomerID',
    values='Quantity',
    aggfunc='sum',
    fill_value=0
)

# Compute cosine similarity between items
item_similarity = cosine_similarity(item_customer_matrix)

# Convert to DataFrame for easier lookup
item_similarity_df = pd.DataFrame(
    item_similarity,
    index=item_customer_matrix.index,
    columns=item_customer_matrix.index
)




item_similarity_df.to_csv("item_similarity_matrix.csv",index=True)

```

```python
item_similarity_df=pd.read_csv("D:/DS ML&AI/Capstone4/env/Scripts/item_similarity_matrix.csv",index_col=0)

item_similarity_df.index = item_similarity_df.index.str.strip()
item_similarity_df.columns = item_similarity_df.columns.str.strip()
    
def recommend_similar_items(product_name, top_n=5):
    """
    Recommend items similar to a given product using cosine similarity.
    """
    if product_name not in item_similarity_df.index:
        return f"Product '{product_name}' not found in dataset."

    #Sort by similarity scores (excluding itself)
    similar_items = item_similarity_df[product_name].sort_values(ascending=False)[1:top_n+1]
    return list(similar_items.index)
```

```python
# Example: Recommend items similar to a product
print(recommend_similar_items("4 PURPLE FLOCK DINNER CANDLES", top_n=5))
```
